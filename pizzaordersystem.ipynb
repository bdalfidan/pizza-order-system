```python
import csv
import datetime

# defining pizza superclass
class Pizza:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost

# defining pizza subclasses
class ClassicItalianPizza(Pizza):
    def __init__(self):
        super().__init__("Classic Italian Pizza", 20)

class TurkishStylePizza(Pizza):
    def __init__(self):
        super().__init__("Turkish Style Pizza", 25)

class AmericanPizza(Pizza):
    def __init__(self):
        super().__init__("American Pizza", 30)

class PlainPizza(Pizza):
    def __init__(self):
        super().__init__("Plain Pizza", 15)

# creating sauce classes
class Sauce:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
    
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost

class BBQSauce(Sauce):
    def __init__(self):
        super().__init__("BBQ Sauce", 3)

class SweetChiliSauce(Sauce):
    def __init__(self):
        super().__init__("Sweet Chili Sauce", 2.5)

class BuffaloSauce(Sauce):
    def __init__(self):
        super().__init__("Buffalo Sauce", 2.5)

class RanchSauce(Sauce):
    def __init__(self):
        super().__init__("Ranch Sauce", 2)

# creating pizza menu
pizza_menu = {
    1: ClassicItalianPizza(),
    2: TurkishStylePizza(),
    3: AmericanPizza(),
    4: PlainPizza()
}

# creating sauce menu
sauce_menu = {
    11: BBQSauce(),
    12: SweetChiliSauce(),
    13: BuffaloSauce(),
    14: RanchSauce(),
}

class PizzaOrder:
    def __init__(self, pizza, sauce):
        self.pizza = pizza
        self.sauce = sauce
    
    def get_description(self):
        return f"{self.pizza.get_description()} with {self.sauce.get_description()}"

    def get_cost(self):
        return self.pizza.get_cost() + self.sauce.get_cost()

def print_menu():
    print("1: Classic Italian - 20 TL")
    print("2: Turkish Style - 25 TL")
    print("3: American - 30 TL")
    print("4: Plain - 15 TL")
    print("* and sauce of your choice:")
    print("11: BBQ Sauce - 3 TL")
    print("12: Sweet Chili Sauce - 2.5 TL")
    print("13: Buffalo Sauce - 2.5 TL")
    print("14: Ranch Sauce - 2 TL")

def get_pizza_choice():
    choice = input("Enter pizza choice (1-4): ")
    while choice not in ["1", "2", "3", "4"]:
        choice = input("Invalid choice. Enter pizza choice (1-4): ")
    if choice == "1":
        return ClassicItalianPizza()
    elif choice == "2":
        return TurkishStylePizza()
    elif choice == "3":
        return AmericanPizza()
    elif choice == "4":
        return PlainPizza()

def get_sauce_choice():
    choice = input("Enter sauce choice (11-14): ")
    while choice not in ["11", "12", "13", "14"]:


    # print menu
     print("Enter pizza choice: ")
    for key, value in pizza_menu.items():
      print(f"{key}: {value.get_description()} - {value.get_cost()} TL")
    print("*" * 40)
    print("Enter sauce choice: ")
    for key, value in sauce_menu.items():
        print(f"{key}: {value.get_description()} - {value.get_cost()} TL")
    print("*" * 40)

    # getting user input for pizza and sauce choices
    pizza_choice = int(input("Enter pizza choice (1-4): "))
    sauce_choice = int(input("Enter sauce choice (11-14): "))

    # calculating total cost
    pizza_cost = pizza_menu[pizza_choice].get_cost()
    sauce_cost = sauce_menu[sauce_choice].get_cost()
    total_cost = pizza_cost + sauce_cost

    # getting user information
    name = input("Name: ")
    id_number = input("Id number: ")
    credit_card_number = input("Credit card number: ")
    credit_card_password = input("Cedit card password: ")

    # writing order information to CSV file
    with open('Orders_Database.csv', mode='a') as order_file:
        order_writer = csv.writer(order_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        order_writer.writerow([name, id_number, credit_card_number, pizza_menu[pizza_choice].get_description(), sauce_menu[sauce_choice].get_description(), total_cost, datetime.datetime.now(), credit_card_password])
```
